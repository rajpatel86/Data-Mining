# Step 1: Load Data

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('OnlineRetail.csv',encoding='ISO-8859-1')
data
```

# Step 2: Drop Missing Invoices or Descriptions

```python
data1=data.copy()
data1.dropna(subset=['InvoiceNo'], inplace=True)
data1
```

```python
data2=data.copy()
data2.dropna(subset=['Description'], inplace=True)
data2
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
data3 = data.copy()
data3 = data3[~data3['InvoiceNo'].str.startswith('C')]
data3
```

# Step 4: Keep Only Positive Quantities

```python
data4 = data.copy()
data4 = data4[data4['Quantity'] > 0]
data4
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
data5 = data.copy()
data5['Description'] = data5['Description'].str.lower().str.replace(' ', '', regex=False)
data5
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
data6 = data.copy()
data6 = data6[data6['Country'] == 'United Kingdom']
data6
```

# Step 7: Remove Duplicates

```python
data7 = data.copy()
data7 = data7.drop_duplicates()
data7
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
data8 = data.copy()
data8['TotalPrice'] = data8['Quantity'] * data8['UnitPrice']
data8
```

# Step 9 : Check for Top 10 Selling Products

```python
data9 = data.copy()
top_products = data9.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

# Step 10 : Identify Most Active Customers

```python
data10 = data.copy()
active_customers_qty = (
    data10.groupby('CustomerID')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(active_customers_qty)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
data11= data.copy()
data11['InvoiceDate'] = pd.to_datetime(data11['InvoiceDate'])
data11['Month'] = data11['InvoiceDate'].dt.month
data11['Hour'] = data11['InvoiceDate'].dt.hour
popular_months = data11['Month'].value_counts().sort_index()
print(popular_months)
popular_hours = data11['Hour'].value_counts().sort_index()
print(popular_hours)


import matplotlib.pyplot as plt

# Plot popular months
popular_months.plot(kind='bar', title='Popular Months')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.show()

# Plot popular hours
popular_hours.plot(kind='bar', title='Popular Hours')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Transactions')
plt.show()
```
