# Step 1: Load Data

```python
import pandas as pd
```

```python
df = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df = df.dropna(subset=['InvoiceNo', 'Description'])
```

```python
df
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df = df[~df['InvoiceNo'].str.startswith('C', na=False)]
```

```python
df
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
```

```python
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df.loc[:, 'Description'] = df['Description'].str.lower().str.replace(' ', '', regex=False)
```

```python
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df_uk = df[df['Country'] == 'United Kingdom']
```

```python
df
```

# Step 7: Remove Duplicates

```python
df = df.drop_duplicates()
```

```python
df
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

# Step 9 : Check for Top 10 Selling Products

```python
df = df.dropna(subset=['Description'])
df = df[df['Quantity'] > 0]
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

# Step 10 : Identify Most Active Customers

```python
df = df.dropna(subset=['CustomerID'])
df = df[df['Quantity'] > 0]

most_active = df.groupby('CustomerID')['Quantity'].sum().sort_values(ascending=False).head(10)

print(most_active)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
import matplotlib.pyplot as plt

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df = df.dropna(subset=['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.month
df['Hour'] = df['InvoiceDate'].dt.hour

popular_months = df['Month'].value_counts().sort_index()
popular_hours = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
popular_months.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Transactions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Step 12: Convert to Basket Format

```python
df = df[df['Quantity'] > 0]
df = df.dropna(subset=['InvoiceNo', 'Description'])

df['Description'] = df['Description'].str.lower().str.strip()

basket_quantity = df.pivot_table(index='InvoiceNo',
                                 columns='Description',
                                 values='Quantity',
                                 aggfunc='sum',
                                 fill_value=0)

basket_binary = basket_quantity.applymap(lambda x: 1 if x > 0 else 0)
basket_quantity
basket_binary
```

# Step:13 : Save Basket Format into CSV file

```python
basket_quantity.to_csv("basket_quantity.csv")
basket_binary.to_csv("basket_binary.csv")
```
